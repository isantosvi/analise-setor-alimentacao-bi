```python
import pandas as pd
import unicodedata
arquivo1 = r"C:\Users\isabella.veras\Desktop\projetodados-main\projeto_dados\ESTABELECIMENTO.csv"
arquivo2 = r"C:\Users\isabella.veras\Desktop\projetodados-main\projeto_dados\CNAE 2(in).csv"
arquivo3 = r"C:\Users\isabella.veras\Desktop\projetodados-main\projeto_dados\MUNICIPIOS 1(in).csv"


df_est = pd.read_csv(arquivo1, encoding="latin-1", sep=";", engine="python")
df_cnae = pd.read_csv(arquivo2, encoding="utf-8-sig", sep=",")
df_municipios = pd.read_csv(arquivo3, encoding="UTF-8", sep=",")
```

```python
df_cnae.columns = df_cnae.columns.str.strip().str.lower()
df_est.columns = df_est.columns.str.strip().str.lower()
df_municipios.columns = df_municipios.columns.str.strip().str.lower()
```

```python
df_est = df_est.dropna(axis=1, how="all")
df_est = df_est.loc[:, ~df_est.columns.str.contains("Unnamed")]
```

```python
df_est["telefone_principal"] = pd.to_numeric(
    df_est["telefone_principal"],
    errors="coerce"
).astype("Int64").astype(str)

df_est["telefone_secundario"] = pd.to_numeric(
    df_est["telefone_secundario"],
    errors="coerce"
).astype("Int64").astype(str)
```

```python
df_est["codigo_municipio"] = df_est["codigo_municipio"].astype("Int64")
df_est['codigo_pais'] = df_est['codigo_pais'].round().astype("Int64")
df_est['cep'] = df_est['cep'].round().astype("Int64")

df_est["data_situacao_especial"] = pd.to_datetime(df_est["data_situacao_especial"], dayfirst=True, errors="coerce")
df_est["data_inicio_atividade"] = pd.to_datetime(df_est["data_inicio_atividade"], dayfirst=True, errors="coerce")
df_est["data_situacao"] = pd.to_datetime(df_est["data_situacao"], dayfirst=True, errors="coerce")
```
