Evolução da empresa

```python
aberturas_por_ano = df_alimentacao.groupby("ano_abertura").size()
print(aberturas_por_ano )
```

```python
# -----------------------------------------
# 4) Crescimento ou declínio nos últimos anos
# -----------------------------------------
def evolucao_abertura_empresas(df):
    base = df.copy()
    base["data_inicio_atividade"] = pd.to_datetime(base["data_inicio_atividade"], errors="coerce")
    base["AnoAbertura"] = base["data_inicio_atividade"].dt.year

    resumo = (
        base.dropna(subset=["AnoAbertura"])
        .groupby("AnoAbertura")
        .size()
        .reset_index(name="QuantidadeEmpresas")
        .sort_values("AnoAbertura")
    )

    resumo["CrescimentoPercentual"] = (
        resumo["QuantidadeEmpresas"].pct_change() * 100
    ).round(2)

    return resumo
```
